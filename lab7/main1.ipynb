```python
import pandas as pd
from matplotlib.pylab import *
import statsmodels.api as sm
```

#### Import data


```python
df = pd.read_csv("data.csv")
df.head()
```

#### Using sm


```python
X = df[["hours", "iq", "attendance"]]
y = df["score"]

# Add constant for intercept
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

model.summary()
```

#### Predict


```python
new_student = {"hours": 5, "iq": 115, "attendance": 0.9}
X_new = pd.DataFrame([new_student])
X_new = sm.add_constant(X_new, has_constant='add')

model.predict(X_new)
```
