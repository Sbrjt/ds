```python
import pandas as pd
from matplotlib.pylab import *
from sklearn.linear_model import LinearRegression
```

#### Import data


```python
df = pd.read_csv("data.csv")
df.head()
```

#### Make model


```python
X = df[["hours", "iq", "attendance"]]
y = df["score"]

model = LinearRegression()
model.fit(X, y)

print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

#### Predict


```python
new_student = {
    "hours": 8,
    "iq": 110,
    "attendance": 85,
}


model.predict(pd.DataFrame([new_student]))
```

#### Graphs


```python
predictors = ["hours", "iq", "attendance"]

for col in predictors:
    scatter(df[col], y)
    xlabel(col)
    ylabel('Score')
    title('Score vs ' + col)
    show()
```
