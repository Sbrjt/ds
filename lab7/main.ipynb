```python
import pandas as pd
from matplotlib.pylab import *
from sklearn.linear_model import LinearRegression
```

#### Import data


```python
df = pd.read_csv("data.csv")
df.head()
```

#### Make model


```python
X = df[["hours", "iq", "attendance"]]
y = df["score"]

model = LinearRegression()
model.fit(X, y)

print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

#### Predict


```python
new_student = {
    "hours": 8,
    "iq": 110,
    "attendance": 85,
}


model.predict(pd.DataFrame([new_student]))
```

#### Graphs


```python
import matplotlib.pyplot as plt
import numpy as np

# Example feature selections
x = df["hours"]
y = df["score"]
color = df["iq"]
size_raw = df["attendance"]

# Scale sizes so they are visually different
size = (size_raw - size_raw.min()) / (
    size_raw.max() - size_raw.min()
) * 300 + 50
# Now size ranges ~ 50 â†’ 350

plt.figure(figsize=(8, 5))
scatter = plt.scatter(x, y, c=color, s=size, alpha=0.7)

plt.xlabel("Hours")
plt.ylabel("Score")
plt.title("Score vs Hours with IQ as Color and Attendance as Size")

# colorbar for IQ
cbar = plt.colorbar(scatter)
cbar.set_label("IQ")

plt.show()
```

```python
predictors = ["hours", "iq", "attendance"]

for col in predictors:
    scatter(df[col], y)
    xlabel(col)
    ylabel('Score')
    title('Score vs ' + col)
    show()
```
