```python
from sklearn.datasets import load_iris
import pandas as pd
from matplotlib.pylab import *
```

```python
iris = load_iris()

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df.head()
```

```python
df.describe()
```

```python
summary = {
    "mean": df.mean(),
    "median": df.median(),
    "mode": df.mode().iloc[0],
    "std_dev": df.std(),
    "range": df.max() - df.min(),
    "iqr": df.quantile(0.75) - df.quantile(0.25),
}

pd.DataFrame(summary)
```

```python
help(boxplot)
```

```python
data = df['petal width (cm)']

boxplot(
    data,
    patch_artist=True,
    notch=True,
    meanline=True,
    showmeans=True,
)


ylabel('Petal Width')
grid(linestyle='--')
```
